```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from google.colab import files
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/desafio/baba/BABA.csv", )
```

```python
dummies_mes = pd.get_dummies(df['mes'])
df['MesAgosto'] = dummies_mes['agosto']
df['MesSetembro'] = dummies_mes['agosto']
df['MesOutubro'] = dummies_mes['outubro']
df['MesNovembro'] = dummies_mes['novembro']
df['MesDezembro'] = dummies_mes['dezembro']
df['MesJaneiro'] = dummies_mes['janeiro']
df['MesFevereiro'] = dummies_mes['fevereiro']
df['MesMarco'] = dummies_mes['marco']
df['MesAbril'] = dummies_mes['abril']
df['MesMaio'] = dummies_mes['maio']
df['MesJunho'] = dummies_mes['junho']
df['MesJulho'] = dummies_mes['julho']
```

```python
df.head()
```

```python
treino = df.iloc[0:365,] 
teste = df.iloc[366:396,]
```

```python
treino
```

```python
teste
```

```python
y_treino = treino['venda']
y_teste = teste['venda']
```

```python
x_treino = treino.loc[:,['MesAgosto','MesSetembro','MesOutubro','MesNovembro','MesDezembro','MesJaneiro','MesFevereiro','MesMarco','MesAbril','MesMaio','MesJunho','MesJulho','margem','desconto','outdesc','outmg']]
```

```python
x_treino
```

```python
x_teste = teste.loc[:,['MesAgosto','MesSetembro','MesOutubro','MesNovembro','MesDezembro','MesJaneiro','MesFevereiro','MesMarco','MesAbril','MesMaio','MesJunho','MesJulho','margem','desconto','outdesc','outmg']]
```

```python
x_teste
```

```python
mod = LinearRegression()
#from sklearn.tree import DecisionTreeRegressor
#mod = DecisionTreeRegressor()
mod.fit(x_treino, y_treino)
print(mod.score(x_treino, y_treino))
```

```python
prev = mod.predict(x_teste)
```

```python
pd.DataFrame(prev).to_csv('predict.csv')
```
